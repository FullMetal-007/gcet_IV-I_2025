```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define the layers and their properties
layers_list = [
    ("Input\n(31 features)", "Dense (32, relu)\nL2=0.01"),
    ("Dense (32, relu)\nL2=0.01", "Dropout 0.3"),
    ("Dropout 0.3", "Dense (16, relu)\nL2=0.01"),
    ("Dense (16, relu)\nL2=0.01", "Dropout 0.3"),
    ("Dropout 0.3", "Dense (1, sigmoid)")
]

# Add edges to the graph
for src, dst in layers_list:
    G.add_edge(src, dst)

# Position nodes vertically
pos = nx.spring_layout(G, seed=42)  # Or use shell_layout for layered style
pos = {node: (0, -i) for i, node in enumerate(G.nodes())}

# Draw the graph
plt.figure(figsize=(6, 8))
nx.draw(G, pos, with_labels=True, node_size=4000,
        node_color="lightblue", font_size=10, font_weight="bold",
        arrowsize=20)

plt.title("Neural Network Architecture", fontsize=14)
plt.axis("off")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define layers (label, color)
layers_info = [
    ("Input\n(31 features)", "skyblue"),
    ("Dense\n32 units\nReLU\nL2=0.01", "lightgreen"),
    ("Dropout\nrate=0.3", "lightcoral"),
    ("Dense\n16 units\nReLU\nL2=0.01", "lightgreen"),
    ("Dropout\nrate=0.3", "lightcoral"),
    ("Dense\n1 unit\nSigmoid", "gold")
]

# Add edges between consecutive layers
for i in range(len(layers_info) - 1):
    G.add_edge(layers_info[i][0], layers_info[i + 1][0])

# Position nodes vertically
pos = {layer[0]: (0, -i) for i, layer in enumerate(layers_info)}

# Colors must match graph nodes order
node_colors = [color for (label, color) in layers_info if label in G.nodes()]

# Draw graph
plt.figure(figsize=(6, 8))
nx.draw(
    G, pos,
    with_labels=True,
    node_size=4500,
    node_color=node_colors,
    font_size=9,
    font_weight="bold",
    arrowsize=20,
    edge_color="gray"
)

plt.title("Sequential Model Architecture", fontsize=14)
plt.axis("off")
plt.show()
```
